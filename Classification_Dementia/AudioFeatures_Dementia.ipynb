Remove the interviewer segments from the audio files and save the segmented audio files

```python
#DEMENTIA

# Check for existing processed audio files
processed_dementia_files = [os.path.join(folder_path, file) for file in os.listdir(folder_path) if file.endswith('_interviewee.wav')]

# Initialize an empty list to store audio files that need processing
dementia_files_to_process = []

# Check if an audio file already has a processed version
for audio_file in audio_files:
    file_name = os.path.splitext(os.path.basename(audio_file))[0]
    processed_file = os.path.join(folder_path, f'{file_name}_interviewee.wav')
    if processed_file not in processed_dementia_files:
        dementia_files_to_process.append(audio_file)

# Process the audio files
for audio_file in dementia_files_to_process:
    try:
        # Load the CSV file with speaker, begin, and end timestamps
        csv_file = os.path.splitext(audio_file)[0] + '.csv'
        if not os.path.exists(csv_file):
            print(f"Skipping {audio_file}: No CSV file found.")
            continue

        df = pd.read_csv(csv_file)

        # Initialize an empty list to store interviewee segments
        interviewee_segments = []

        # Iterate through the rows of the DataFrame
        for _, row in df.iterrows():
            if row['speaker'] == 'PAR':
                # Load the audio file
                audio = AudioSegment.from_wav(audio_file)

                # Extract the segment based on begin and end timestamps
                begin_time = row['begin'] * 1000  # Convert to milliseconds
                end_time = row['end'] * 1000  # Convert to milliseconds
                segment = audio[begin_time:end_time]

                # Add the segment to the list
                interviewee_segments.append(segment)

        if len(interviewee_segments) == 0:
            print(f"Skipping {audio_file}: No interviewee segments found.")
            continue

        # Concatenate the interviewee segments into a single audio file
        final_audio = sum(interviewee_segments)

        # Save the final audio file if it doesn't already exist
        output_file = os.path.join(folder_path, f'{os.path.splitext(os.path.basename(audio_file))[0]}_interviewee.wav')
        if not os.path.exists(output_file):
            final_audio.export(output_file, format='wav')
            print(f"Processed {audio_file} and saved as {output_file}")
        else:
            print(f"Skipping {audio_file}: Processed audio file already exists.")
    except Exception as e:
        print(f"Error processing {audio_file}: {str(e)}")  
    
    
#CONTROL

# Check for existing processed audio files
processed_control_files = [os.path.join(folder_path2, file) for file in os.listdir(folder_path2) if file.endswith('_interviewee.wav')]

# Initialize an empty list to store audio files that need processing
control_files_to_process = []

# Check if an audio file already has a processed version
for audio_file in audio_files2:
    file_name = os.path.splitext(os.path.basename(audio_file))[0]
    processed_file = os.path.join(folder_path2, f'{file_name}_interviewee.wav')
    if processed_file not in processed_control_files:
        control_files_to_process.append(audio_file)

# Process the audio files
for audio_file in control_files_to_process:
    try:
        # Load the CSV file with speaker, begin, and end timestamps
        csv_file = os.path.splitext(audio_file)[0] + '.csv'
        if not os.path.exists(csv_file):
            print(f"Skipping {audio_file}: No CSV file found.")
            continue

        df2 = pd.read_csv(csv_file)

        # Initialize an empty list to store interviewee segments
        interviewee_segments = []

        # Iterate through the rows of the DataFrame
        for _, row in df2.iterrows():
            if row['speaker'] == 'PAR':
                # Load the audio file
                audio = AudioSegment.from_wav(audio_file)

                # Extract the segment based on begin and end timestamps
                begin_time = row['begin'] * 1000  # Convert to milliseconds
                end_time = row['end'] * 1000  # Convert to milliseconds
                segment = audio[begin_time:end_time]

                # Add the segment to the list
                interviewee_segments.append(segment)

        if len(interviewee_segments) == 0:
            print(f"Skipping {audio_file}: No interviewee segments found.")
            continue

        # Concatenate the interviewee segments into a single audio file
        final_audio = sum(interviewee_segments)

        # Save the final audio file if it doesn't already exist
        output_file = os.path.join(folder_path2, f'{os.path.splitext(os.path.basename(audio_file))[0]}_interviewee.wav')
        if not os.path.exists(output_file):
            final_audio.export(output_file, format='wav')
            print(f"Processed {audio_file} and saved as {output_file}")
        else:
            print(f"Skipping {audio_file}: Processed audio file already exists.")
    except Exception as e:
        print(f"Error processing {audio_file}: {str(e)}")
```

## Feature Extraction

```python
# Function to calculate RMSE
def calculate_rmse(y):
    rmse = np.sqrt(np.mean(np.square(y)))
    return rmse

# Function to extract audio features
def extract_audio_features(audio_file):
    # Load the audio file using librosa
    y, sr = librosa.load(audio_file)

    # Calculate RMSE
    log_energy = calculate_rmse(y)

    # Extract time-dependent features using librosa
    mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)
    gtccs = librosa.feature.chroma_cens(y=y, sr=sr)
    delta_gtccs = librosa.feature.delta(gtccs)
    formants = librosa.effects.harmonic(y=y)
    fundamental_frequency, _, _ = librosa.pyin(y, fmin=librosa.note_to_hz('C2'), fmax=librosa.note_to_hz('C7'))

    return {
        'MFCCs': mfccs,
        'GTCCs': gtccs,
        'Delta_GTCCs': delta_gtccs,
        'Log_Energy': log_energy,
        'Formants': formants,
        'Fundamental_Frequency': fundamental_frequency,
        'sr': sr  # Change the key to 'sr' to match the variable name
    }
```

```python
all_features = []

#Iterate over audio files and extract features
for audio_file in audio_files:
    features = extract_audio_features(audio_file)
    all_features.append(features)

#     # Visualize the waveform and MFCCs (optional)
#     plt.figure(figsize=(10, 6))
#     librosa.display.waveshow(features['MFCCs'], sr=features['sr'])  
#     plt.title('MFCCs')
#     plt.show()
```
